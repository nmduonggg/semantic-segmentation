```python
import albumentations as A
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/home/nguyenduong/Projects/semantic-segmentation/assests/image_labels/Seq05VD_f05100.png')
plt.imshow(img)
```

```python
mask1 = np.zeros((img.shape[0], img.shape[1]))
mask1[:img.shape[0]//3] = 1

mask2 = np.zeros((img.shape[0], img.shape[1]))
mask2[img.shape[0]//3: img.shape[0]//3*2]=1
```

```python
plt.imshow(mask1)
```

```python
plt.imshow(mask2)
```

```python
transform = A.Compose([
    A.VerticalFlip(p=0.5),
    A.Rotate(limit=30, p=1)
])

tfs = transform(image=img, masks=[mask1, mask2])
new_img = tfs['image']
nmask1, nmask2 = tfs['masks']
```

```python
plt.imshow(new_img)
```
